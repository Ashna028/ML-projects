### <center>Task 1: Exploring and Visualizing the Iris Dataset</center>
---


```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

---

```python
# Load the built-in Iris dataset
df = sns.load_dataset("iris")
```

---

```python
# Print dataset shape (rows, columns)
print("Shape of dataset:", df.shape)
```

---

```python
# Print column names
print("Column names:", df.columns)
```

---

```python
# Show first few rows of the dataset
print(df.head())
```

---

```python
# Display data types and non-null info
print(df.info())
```

---

```python
# Summary statistics of numeric features
print(df.describe())
```

---

```python
# Pairplot to visualize feature relationships by species
sns.pairplot(df, hue="species")
plt.suptitle("Pairwise Feature Relationships by Species", y=1.02)
plt.show()
```

---

```python
# Histograms for distribution of each feature
df.hist(figsize=(10, 8), edgecolor='black')
plt.suptitle("Histogram of Iris Features", y=1.02)
plt.show()
```

---

```python
# Boxplots to identify outliers
sns.boxplot(data=df)
plt.title("Boxplot of Iris Dataset Features")
plt.show()
```
